```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
def load_data():
    data=pd.read_csv('/content/drive/MyDrive/Perceptron/dataset_LP_2.csv',header=None)
    data=np.asmatrix(data,dtype='float64')
    np.random.seed(0)
    np.random.shuffle(data)
    #print(data)
    #print(data.shape)
    train,test=np.split(data,[int(0.7*data.shape[0])])
    #print(train)
    #print(train.shape)
    #print(test)
    #print(test.shape)
    return train,test;
```

```
def load_data1():
    data=pd.read_csv("/content/drive/MyDrive/Perceptron/dataset_LP_1.txt",header=None)
    data=np.asmatrix(data,dtype='float64')
    #print(data)
    np.random.seed(0)
    np.random.shuffle(data)
    train,test=np.split(data,[int(0.7*data.shape[0])])
    return train,test;
```

```
def linear_perceptron(train):
    #print(train)
    features=train[:,:-1]
    labels=train[:,-1]
    #print(features)
    #print(labels)
    w=np.zeros(shape=(1,features.shape[1]+1))

    #misclassified=[]
    for i in range(int(pow(10,3))):
        num_misclassified=0
        for j in range(features.shape[0]):
            #curr_features=features[j]
            curr_features=np.insert(features[j],0,1)
            #print(features[j].shape)
            #print(w.shape)
            #curr_features=curr_features.transpose()
            val=np.dot(w,curr_features.transpose())
            #print(val)
            if val>0:
                t=1.0
            else:
                t=0.0
            #delta=labels[j]-t
            if t!=labels[j]:
                num_misclassified+=1
                w+=0.01*((labels[j]-t)*curr_features)
        #print(num_misclassified)
        if num_misclassified==0:
            #print("yayyyy")
            break
    return w
```

```
def find_accuracy(w,test):
    features=test[:,:-1]
    labels=test[:,-1]
    num_misclassified=0
    for i in range(features.shape[0]):
        #print(features[0].shape)
        #print(w.shape)
        curr_features=np.insert(features[i],0,1)
        #print(curr_features.shape)
        val=np.dot(w,curr_features.transpose())
        #print(val)
        if val>0:
            t=1.0
        else:
            t=0.0
        if t!=labels[i]:
            num_misclassified+=1
    #print(num_misclassified)
    return 1-(num_misclassified/features.shape[0])
```

```
#main
#dataset1
train,test=load_data1()
w=linear_perceptron(train)
accuracy=find_accuracy(w,test)
print("Accuracy with first data set is",accuracy)

#dataset2
train,test=load_data()
w=linear_perceptron(train)
accuracy=find_accuracy(w,test)
print("Accuracy with second data set is",accuracy)
```
